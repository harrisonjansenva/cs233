## Data Analsyis: Child Race Data

### a. Load Data

<u>**Question 1. (pt 2)**</u> :

Load data from the file *children_raw.csv* into a *Table* called `raw_child_data`. 

```python
raw_child_data = pd.read_csv("children_raw.csv").set_index('State')
raw_child_data
```

<u>**Question 2. (pt 2)**</u> :

Create new table called `child_data` that uses data from `raw_child_data` to calculate what fraction each race represents in the child population for each state. The columns in the new table should be called `Fraction White`, `Fraction Black`, etc. 

```python
child_data = raw_child_data.drop(columns=["Total"]).div(raw_child_data['Total'], axis=0)
child_data.columns = ['Fraction ' + col for col in child_data.columns]
child_data
```

### b. Analyze Data

Draw a barchart with the fraction of the child populations that are hispanic.

```python
child_data['Fraction Hispanic'].plot(kind='bar', figsize=(14,6), ylabel='Percentage', title="Percentage of Hispanic Children, by State")
```
