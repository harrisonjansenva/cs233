A **great circle** or orthodrome is the circular intersection of a sphere and a plane passing through the sphere's center point. It displays the ***shortest*** route between two points on the sphere.

**Let's do the caculation using Python**


Compose a program that 

 - assign four floats **lat1, lon1, lat2, and lon2**—(the latitude and longitude, in **degrees**, of two points on the earth)
 - writes the great-circle distance between them using the great-circle distance **d** given by the following equation:

```
d = 6371 arccos[sin(lat1) sin(lat2) + cos(lat1) cos(lat2) cos(lon1 − lon2)]
```

Use your program to compute the great-circle distance between Paris (48.8575° N, 2.3514° E) and Boise (43.6150° N, 116.2023° W).

- Read references on the math package at: https://docs.python.org/3/library/math.html.
- You need mathematical functions like `math.sin(), math.cos(), math.acos()`.
- The inputs are degrees, whereas Python’s trigonometric functions use **radians**. For example, let's study math.sin(x) function, the reference page https://docs.python.org/3/library/math.html#math.sin metioned: 
"Return the sine of x **radians**". **So, you may need to use `math.radians()` and `math.degrees()` to convert between the two.**


````python
# First, let's difine a function to compute the great circle distance

def great_circle(lat1: float, lon1: float, lat2: float, lon2: float) -> float:
    """
    Calculate the great-circle distance between two points on the Earth's surface.

    Parameters:
    lat1 (float): Latitude of the first point in degrees.
    lon1 (float): Longitude of the first point in degrees.
    lat2 (float): Latitude of the second point in degrees.
    lon2 (float): Longitude of the second point in degrees.

    Returns:
    float: The great-circle distance between the two points in kilometers.
    """
    
    # Convert latitude and longitude from degrees to radians
    rlat1 = math.radians(lat1)
    rlat2 = math.radians(lat2)
    rlon1 = math.radians(lon1)
    rlon2 = math.radians(lon2)

    # Calculate the great-circle distance
    d = 6371 * math.acos(math.sin(rlat1) * math.sin(rlat2) + math.cos(rlat1) * math.cos(rlat2) * math.cos(rlon1 - rlon2))
    
    return d
````

````python
# Second, let's computer d between Paris and Boise

lat1 = 48.8575
lon1 = 2.3514
lat2 = 43.6150
lon2 = -116.2023

distance = great_circle(lat1, lon1, lat2, lon2)

# Use F-strings tor formated output
print(f'Latitude and longitude of first point: {lat1}, {lon1}°')
print(f'Latitude and longitude of the second point: {lat2}°, {lon2}°')
print(f'Great-circle distance between the two points: {distance:.3f} km')
````
